```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
from scipy.stats import linregress
```

```python
data = pd.read_csv('casos.csv',index_col=0,parse_dates=True)
data.dropna(inplace=True) # cerciorate de no tener NaNs
data
```

```python
# Para el caso 0.01
Ti_s = data.Tcub

fig, ax = plt.subplots(figsize=(6,6))


x = np.linspace(-10,10,2)
y = x 

X = np.array( data.Ti-data.Ti.mean()   )
Y = np.array( Ti_s-data.Ti.mean() )
m,b,r_value, p_value,std_err = linregress(X,Y)

ax.scatter(X,Y)
ax.plot(x,y,'k-') 
ax.plot(x,x*m+b,'r-') 


ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.set_box_aspect(1)
ax.set_ylabel('Ti_s [Â°C]')
ax.set_xlabel('Ti   [Â°C]')
ax.grid()
print(m,b,r_value)
```
